```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('dark_background')
```

# Data Cleaning

## Reading CSV

```python
df = pd.read_csv('../data/zomato.csv') # https://www.kaggle.com/datasets/himanshupoddar/zomato-bangalore-restaurants/
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1, inplace=True)
df.head()
```

```python
df.info()
```

## Dropping duplicates

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

## Removing "NEW", "-", "/5" from rate

```python
def handlerate(value):
    return float(str(value).split('/')[0]) if value not in ['NEW', '-'] else np.nan
```

```python
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

## Filling Null Values in Rate column with Mean

```python
df['rate'].fillna(df['rate'].mean(), inplace=True)
df.rate.isnull().sum()
```

```python
df.info()
```

## Dropping Null Values

```python
df.dropna(inplace=True)
df.head()
```

```python
df.rename(columns={"approx_cost(for two people)":"cost2plates", "listed_in(type)": "type"}, inplace=True)
df.head()
```

```python
df.location.unique()
```

```python
df['listed_in(city)'].unique()
```

## Listed in(city) and location, both are there, lets keeps only one

```python
df.drop(['listed_in(city)'], axis=1, inplace=True)
df.head()
```

```python
df.cost2plates.unique()
```

```python
df.cost2plates = df.cost2plates.apply(lambda x: float(x.replace(',', '')))
df.head()
```

```python
df.cost2plates.unique()
```

## Cleaning Rest Type Column 

```python
rest_types = df.rest_type.value_counts(ascending=False)
rest_types
```

```python
rest_types_less1000 = rest_types[rest_types < 1000]
rest_types_less1000
```

## Making Rest Types than 1000 in frequency as other

```python
df['rest_type'] = df['rest_type'].apply(lambda x: x if x not in rest_types_less1000 else 'others')
df['rest_type'].value_counts()
```

## Cleaning location column

```python
location = df['location'].value_counts(ascending=False)
location_less300 = location[location < 300]
location_less300
```

```python
df['location'] = df['location'].apply(lambda x: x if x not in location_less300 else 'others')
df['location'].value_counts()
```

```python
df.head()
```

## Cleaning Cuisines Column 

```python
cuisines = df['cuisines'].value_counts(ascending=False)
cuisines_less100 = cuisines[cuisines < 100]
cuisines_less100
```

```python
df['cuisines'] = df['cuisines'].apply(lambda x: x if x not in cuisines_less100 else 'others')
df['cuisines'].value_counts()
```

# Visualization

## Count Plot of Various Location

```python
plt.figure(figsize=(16,10))
ax = sns.countplot(df, x="location")
plt.xticks(rotation=90);
```

## Visualizing Online Order

```python
plt.figure(figsize=(6,6))
sns.countplot(df, x='online_order', palette='inferno')
```

## Visualizing Book Table

```python
plt.figure(figsize=(6,6))
sns.countplot(df, x='book_table', palette='rainbow')
```

## Visualizing Online Order vs Rate

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order', y='rate', data=df)
```

## Visualizing Book Table vs Rate

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='book_table', y='rate', data=df)
```

## Visualizing Online Order Facility, Location Wise

```python
df1 = pd.DataFrame(df.groupby(['location', 'online_order'])['name'].count())
df1 = pd.pivot_table(df1, 
                     values=None, 
                     index=['location'], 
                     columns=['online_order'], 
                     fill_value=0,
                     aggfunc=np.sum)
df1
```

```python
df1.plot(kind='bar', figsize=(15,8))
```

## Visualizing Book Table Facility, Location Wise

```python
df1 = pd.DataFrame(df.groupby(['location', 'book_table'])['name'].count()).rename(columns={"name": "count_of_rest"})
df1 = pd.pivot_table(df1,
                     values=None,
                     index=['location'],
                     columns=['book_table'],
                     fill_value=0,
                     aggfunc=np.sum)
df1
```

```python
df1.plot(kind='bar', figsize=(15,8))
```

## Visualizing Types of Restaurants vs Rate

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='type', y='rate', data=df, palette='inferno')
plt.xticks(rotation=45);
```

## Grouping Types of Restaurants, location wise

```python
df3 = pd.DataFrame(df.groupby(['location', 'type'])['name'].count()).rename(columns={"name": "count_of_rest"})
df3 = pd.pivot_table(df3,
                     values=None,
                     index=['location'],
                     columns=['type'],
                     fill_value=0,
                     aggfunc=np.sum)
df3
```

```python
df3.plot(kind='bar', figsize=(36,8))
```

## No. of Votes, Location Wise

```python
df4 = df[['location', 'votes']]
df4.drop_duplicates()
df5 = df4.groupby(['location'])['votes'].sum()
df5 = df5.to_frame()
df5 = df5.sort_values('votes', ascending=False)
df5.head()
```

```python
plt.figure(figsize=(15,8))
sns.barplot(df5, x=df5.index, y=df5['votes'])
plt.xticks(rotation=90);
```

## Visualizing Top Cuisines

```python
df6 = df[['cuisines', 'votes']]
df6.drop_duplicates()
df7 = df6.groupby(['cuisines'])['votes'].sum()
df7 = df7.to_frame()
df7 = df7.sort_values('votes', ascending=False)

df7 = df7.iloc[1:, :]

df7.head()
```

```python
plt.figure(figsize=(15,8))
sns.barplot(df7, x=df7.index, y=df7['votes'])
plt.xticks(rotation=90);
```
